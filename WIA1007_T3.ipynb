```python
import numpy as np

# 1. Create a numpy array of size 2 x 2
a=np.array([[1,2],[3,4]])
print (a)
```

```python
#2. Create a random numpy array of size 3 x 4

b=np.random.rand(3,4)
print (b)
```

```python
#3. Create a numpy array of ones and a numpy array of zeroes

c=np.full((3,3),1)

d=np.full((4,3),0)

print("array of ones: \n",c)
print("array of zeroes: \n",d)
```

```python
#4. Create a random numpy array of size 6x3 and transpose it

e=np.random.randint(10,50,size=(6,3))

print("original 6x3 array: \n",e)

e_tranpose=np.transpose(e)
print("transposed 3x6 array: \n",e_tranpose)
```

```python
#5. Create two random numpy array of size n x p and p x m and then compute the dot product

n,p,m=2,3,4

a1=np.random.randint(10,50,size=(n,p))
b1=np.random.randint(10,50,size=(p,m))

print("array 1: \n",a1)
print("array 2: \n",b1)

result=np.dot(a1,b1)
print("dot product: \n",result)
```
